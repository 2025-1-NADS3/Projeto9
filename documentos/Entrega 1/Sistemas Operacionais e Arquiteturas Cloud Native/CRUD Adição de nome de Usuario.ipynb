```python
# Lista para armazenar os dados
dados = []

def criar():
    item = input("Digite o Seu Nome Completo: ")
    dados.append(item)
    print("Item criado com sucesso!")

def ler():
    if len(dados) == 0:
        print("Nenhum dado disponível.")
    else:
        print("Itens cadastrados:")
        for i, item in enumerate(dados):
            print(f"ID {i}: {item}")

def atualizar():
    ler()
    try:
        id = int(input("Digite o ID do item que deseja atualizar: "))
        if 0 <= id < len(dados):
            novo_item = input("Digite o novo valor: ")
            dados[id] = novo_item
            print("Item atualizado com sucesso!")
        else:
            print("ID inválido.")
    except ValueError:
        print("Por favor, insira um número válido.")

def deletar():
    ler()
    try:
        id = int(input("Digite o ID do item que deseja deletar: "))
        if 0 <= id < len(dados):
            dados.pop(id)
            print("Item deletado com sucesso!")
        else:
            print("ID inválido.")
    except ValueError:
        print("Por favor, insira um número válido.")

def menu():
    while True:
        print("\nEscolha uma opção:")
        print("1. Criar")
        print("2. Ler")
        print("3. Atualizar")
        print("4. Deletar")
        print("5. Sair")

        opcao = input("Digite a opção desejada: ")
        if opcao == "1":
            criar()
        elif opcao == "2":
            ler()
        elif opcao == "3":
            atualizar()
        elif opcao == "4":
            deletar()
        elif opcao == "5":
            print("Encerrando o programa...")
            break
        else:
            print("Opção inválida, tente novamente.")

# Iniciar o programa
menu()
```
